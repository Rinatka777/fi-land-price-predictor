```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../data/processed/finnish_land_clean.csv")
df["year"] = df["quarter"].str[:4].astype(int)
df["q_num"] = df["quarter"].str[-1].astype(int)

series = df["neliöhinta"]

plt.hist(series, bins=20, edgecolor="black")
sns.histplot(data=df, x="neliöhinta", bins=20, kde=True)

plt.scatter(df["reaalihintaindeksi"], df["neliöhinta"])
sns.scatterplot(data=df, x="reaalihintaindeksi", y="neliöhinta")

df.corr(numeric_only=True)
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")

plt.plot(df["quarter"], df["neliöhinta"], label="Price")
plt.plot(df["quarter"], df["reaalihintaindeksi"], label="Index")
plt.legend()



print(df.head())
print(df.info())
print(df.isnull().sum())
print(df.shape)
print(df.describe())
```
